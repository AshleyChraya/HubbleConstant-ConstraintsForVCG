```python
omega_r0=0.0000245 #Radiation
omega_b0=0.02 #Baryonic matter
H0=69.8 #Hubble constant in (Km/s)/Mpc
```

Defining the function to find luminosity distance and followed by calculating the luminosity distance through integration by quad function of the scipy module for different combination of Omega M and n and from that the distance modulus.

Friedmann equation: $H^2=\Omega_{ch0} H_0^2 (1+z)^4 X^2(z)$ where $X^{2}(z)=\frac{\Omega_{r 0}}{1-\Omega_{r 0}-\Omega_{b 0}}+\frac{\Omega_{b 0} }{1-\Omega_{r 0}-\Omega_{\theta 0}(1+z)}+\frac{\left(\Omega_{m}(1+z)^6+(1-\Omega_{m})(1+z)^n\right)^{1 / 2}}{(1+z)^4}$ 

Calculating the luminosity distance by $d_l = \frac{c(1+z)}{H0} \int_{0}^{z} \frac{dz}{(1+z)^2X(z)}$

```python
def int_func(redshift,omega_m, n):
      return 1/((pow((1+redshift),2)*(omega_r0+(omega_b0/(1+redshift))
      +((1-omega_b0-omega_r0)*((omega_m*(1+redshift)**6)
      +(1-omega_m)*(1+redshift)**n)**(0.5)/(1+redshift)**(4)))**(0.5)))
```

```python
dmthSN, dlthSN, dldataSN, error, dlerror  = (
    np.zeros(len(z)), np.zeros(len(z)), np.zeros(len(z)),np.zeros(len(z)), np.zeros(len(z)))
# From minimization of chi square
omega_m=0.11
n=1.26
c=3*10**5 #Km/s
# DL theoretical from VCG model
for a,redshift in zip(sort_data_SN.index,sort_data_SN.Z_SN):
  g = quad(int_func,0,redshift, args=(omega_m, n))
  dlthSN[a]+=(1+redshift)*g[0]*(c/(H0)) # g[0] because g[1] gives absolute error in the result obtained
# DM from DL theoretical model VCG
for a,i in zip(sort_data_SN.index,dlthSN):
  dmthSN[a]+=5*log(i,10)+25 #m-M=mu=5log(dl/dl,0) where M is the magnitude of the 
#standard candle at some nearby distance dL,0.n the conventional definition of the 
#distance modulus, dL,0 = 10 pc and M at this distance is usually referred to as the
#absolute magnitude. Converting dl,0 in Mpc, we get mu=5log(dl) + 25

# Distance luminosity from data in Mpc
for a,i in zip(sort_data_SN.index,sort_data_SN.DMdata_SN):
  dldataSN[a]+=10**((i/5)+1)/10**6
#10**((i/5)+1) will give in dl in pc or 10**((i/5)-5) in Mpc
# Calculating error bars of DL
for (a,i,j) in zip(sort_data_SN.index,mu,errdm):   
  error[a]+= 10**((i+j)/5+1)/10**6
for (a,i,j) in zip(sort_data_SN.index,dldataSN,error):
    dlerror[a]+=j-i   #Error in data distance luminosity
sort_data_SN["DMth_SN"]=dmthSN
sort_data_SN["dLth_SN"] = dlthSN
sort_data_SN["dLdata_SN"] = dldataSN
sort_data_SN["Error_DL_SN"] = dlerror
%time
```
